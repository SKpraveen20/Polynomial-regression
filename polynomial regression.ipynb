```python
import numpy as np
import pandas as pd
import sklearn.metrics as mt
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data=pd.read_csv("Position_Salaries.csv")
x=data.iloc[:,1:2].values
y=data.iloc[:,2].values
data
```

```python
plt.figure(figsize=(20,8))
plt.scatter(data["Level"],data["Salary"],color="r")
plt.xlabel("Position")
plt.ylabel("Salary")
plt.title("SALARY PREDICTION")
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
poly=PolynomialFeatures(degree=4)
xp=poly.fit_transform(x)
reg=LinearRegression()
reg.fit(xp,y)
```

```python
plt.figure(figsize=(20,8))
plt.scatter(x, y, color = 'blue')
plt.plot(x, reg.predict(xp), color = 'red')
plt.title('Polynomial Regression')
plt.xlabel('Position')
plt.ylabel('Salary')
```

```python
y_pred= reg.predict(xp)
```

```python
print("             MSE : ",(mt.mean_squared_error(y, y_pred)))
print("        R2_Score : ",(mt.r2_score(y, y_pred)))
print("            RMSE : ",(np.sqrt(mt.mean_squared_error(y,y_pred))))
```
